```python
# Projeto 1  Anlise Exploratria de Dados (EDA)

# Objetivo: Analisar vendas da Black Friday utilizando pandas, seaborn e matplotlib

# Importao das bibliotecas necessrias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurao padro para grficos
sns.set(style="whitegrid")

# 1. Leitura do Dataset
df = pd.read_csv('/Users/tamargenzgaulke/Documents/GitHub/Projeto1_EDA/train.csv')

# 2. Viso geral dos dados
print(df.shape)
df.head()

# 3. Limpeza inicial
print(df.isnull().sum())
df = df.dropna() # Removendo linhas com valores ausentes

# 4. Anlise descritiva
print(df.describe())

# 5. Visualizaes principais

# Compras por faixa etria
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Age', order=sorted(df['Age'].unique()))
plt.title('Quantidade de compras por faixa etria')
plt.xticks(rotation=0)
plt.show()

# Compras por gnero
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender')
plt.title('Compras por Gnero')
plt.show()

# Compras por categoria de produto
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Product_Category_1')
plt.title('Compras por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.show()

# Distribuio do valor da compra
plt.figure(figsize=(8,5))
sns.histplot(df['Purchase'], bins=30, kde=True)
plt.title('Distribuio do Valor das Compras')
plt.xlabel('Valor (R$)')
plt.show()

# 6. Insights rpidos
# Exemplo: grupos que mais compram, ticket mdio, categorias com maior volume
# - Usurios entre 26 e 35 anos foram os que mais compraram.
# - Categoria 1 foi a mais vendida.
# - A maioria das compras ficou abaixo de R$ 10.000.


# 7. Concluso
# Escreva suas concluses finais aqui.


```
