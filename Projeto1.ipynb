```python
# Projeto 1 – Análise Exploratória de Dados (EDA)

# Objetivo: Analisar vendas da Black Friday utilizando pandas, seaborn e matplotlib

# Importação das bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração padrão para gráficos
sns.set(style="whitegrid")

# 1. Leitura do Dataset
df = pd.read_csv('/Users/tamargenzgaulke/Documents/GitHub/Projeto1_EDA/train.csv')

# 2. Visão geral dos dados
print(df.shape)
df.head()

# 3. Limpeza inicial
print(df.isnull().sum())
df = df.dropna() # Removendo linhas com valores ausentes

# 4. Análise descritiva
print(df.describe())

# 5. Visualizações principais

# Compras por faixa etária
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Age', order=sorted(df['Age'].unique()))
plt.title('Quantidade de compras por faixa etária')
plt.xticks(rotation=0)
plt.show()

# Compras por gênero
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender')
plt.title('Compras por Gênero')
plt.show()

# Compras por categoria de produto
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Product_Category_1')
plt.title('Compras por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.show()

# Distribuição do valor da compra
plt.figure(figsize=(8,5))
sns.histplot(df['Purchase'], bins=30, kde=True)
plt.title('Distribuição do Valor das Compras')
plt.xlabel('Valor (R$)')
plt.show()

# 6. Insights rápidos
# Exemplo: grupos que mais compram, ticket médio, categorias com maior volume
# - Usuários entre 26 e 35 anos foram os que mais compraram.
# - Categoria 1 foi a mais vendida.
# - A maioria das compras ficou abaixo de R$ 10.000.


# 7. Conclusão
# Escreva suas conclusões finais aqui.


```
